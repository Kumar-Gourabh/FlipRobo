# 1. Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
def wiki(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    titles = soup.find_all(['h1','h2','h3','h4','h5','h6'])
    print('Header tags :', *titles, sep='\n\n')
```

# 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
def imdb(url):
    Names=[]
    Ratings=[]
    Years=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    name=soup.find_all('td',class_="titleColumn")
    for i in name:
        full_name=i.find('a').get_text()
        Names.append(full_name)
    rating=soup.find_all('td',class_="ratingColumn imdbRating")
    for i in rating:
        Ratings.append(i.text.replace('\n',''))
    year=soup.find_all('span',class_="secondaryInfo")
    for i in year:
        Years.append(i.text.replace('(','').replace(')',''))
    import pandas as pd
    movies=pd.DataFrame({})
    movies['Names']=Names
    movies['Ratings']=Ratings
    movies['Years']=Years
    movies=movies[:100]
    print(movies)
    movies.to_csv('Top_100_imdb_movies.csv')
```

```python
imdb('https://www.imdb.com/chart/top/')
```

# 4. Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
def book_page(url):
    book_names=[]
    author_names=[]
    genres=[]
    reviews=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    books=soup.find_all('h4',class_='italic')
    for i in books:
        book_names.append(i.find('a').text)
    author=soup.find_all('p',class_='sans bold')
    for i in author:
        author_names.append(i.text.replace('\n',''))
    genre=soup.find_all('p',class_='genre-links hidden-phone')
    for i in genre:
        genres.append(i.find('a').text)
    review=soup.find_all('p',class_='excerpt')
    for i in review:
        reviews.append(i.text.replace('\n',''))
    import pandas as pd
    books=pd.DataFrame({})
    books['book_names']=book_names
    books['author_names']=author_names
    books['genres']=genres
    books['reviews']=reviews
    books=books.head()
    return(books)
```

```python
df=book_page('https://bookpage.com/reviews')
df
```

# 5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. 
You have to scrape: i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. ii) Top 10 ODI Batsmen in men along with the records of their team and rating. iii) Top 10 ODI bowlers along with the records of their team and rating.

(i) Top 10 ODI Teams

```python
def top_10_teams(url):
    teams=[]
    matches=[]
    points=[]
    ratings=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    team=soup.find_all('span',class_='u-hide-phablet')
    for i in team:
        teams.append(i.text)
    teams=teams[:10]
    
    match_top_row=soup.find_all('td',class_='rankings-block__banner--matches')
    #I observed that top row style is different and remaining rows are different
    for i in match_top_row:
        matches.append(i.text)
    # For remaining 9 teams
    match=soup.find_all('td',class_='table-body__cell u-center-text')
    # I am using even indices only for Matches. Will use odd ones for Points.
    for i in range(0,len(match),2):
        matches.append(match[i].text)
    matches=matches[:10]
    
    point_top_row=soup.find_all('td',class_='rankings-block__banner--points')
    #Similarly, I observed that top row style is different and remaining rows are different
    for i in point_top_row:
        points.append(i.text)
    # For remaining 9 teams
    point=soup.find_all('td',class_='table-body__cell u-center-text')
    # I am using odd indices only for Matches. Will use odd ones for Points.
    for i in range(1,len(match),2):
        points.append(point[i].text)
    points=points[:10]
    
    rating=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
    #Similarly, I observed that top row style is different and remaining rows are different
    for i in rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    # For remaining 9 teams
    rating=soup.find_all('td',class_='table-body__cell u-text-right rating')
    for i in rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    ratings=ratings[:10]
    
    
    import pandas as pd
    teams_odi=pd.DataFrame({})
    teams_odi['teams']=teams
    teams_odi['matches']=matches
    teams_odi['points']=points
    teams_odi['ratings']=ratings
    return(teams_odi)
```

```python
df=top_10_teams('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
df
```

(ii) Top 10 ODI Batsmen

```python
def top_10_players(url):
    players=[]
    teams=[]
    ratings=[]
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    # First Row
    player=soup.find_all('div',class_='rankings-block__banner--name-large')
    for i in player:
        players.append(i.text)
    #Next 9 rows
    player=soup.find_all('td',class_='table-body__cell rankings-table__name name')
    for i in player:
        players.append(i.find('a').text)
    players=players[:10]
    
    # First Row
    team=soup.find_all('div',class_='rankings-block__banner--nationality')
    for i in team:
        teams.append(i.text.replace('\n',''))
    # Next 9 Rows
    team=soup.find_all('span',class_='table-body__logo-text')
    for i in team:
        teams.append(i.text.replace('\n',''))
    teams=teams[:10]
    
    # First Row
    rating=soup.find_all('div',class_='rankings-block__banner--rating')
    for i in rating:
        ratings.append(i.text)
    # Next 9 rows
    rating=soup.find_all('td',class_='table-body__cell rating')
    for i in rating:
        ratings.append(i.text)
    ratings=ratings[:10]
    
    
    import pandas as pd
    batsmen_odi=pd.DataFrame({})
    batsmen_odi['players']=players
    batsmen_odi['teams']=teams
    batsmen_odi['ratings']=ratings
    return(batsmen_odi)
```

# 7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```python
def amazon(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    products=[]
    product=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
    for i in product:
        products.append(i.text)
        
    prices=[]
    price=soup.find_all('span',class_='a-price-whole')
    for i in price:
        prices.append(i.text)
        
    images=[]
    image=soup.find_all('div',class_='a-section aok-relative s-image-fixed-height')
    for i in image:
        images.append(i.find('img'))
    
    ratings=[]
    avg_ratings=[]
    rating=soup.find_all('span',class_='a-icon-alt')
    for i in rating:
        ratings.append(i.text)
    # Need to skip first four ratings as they belong to something else
    ratings=ratings[4:]
    # Extracting first 3 characters of rating
    for j in ratings:
        avg_ratings.append(j[0:3])
       
    import pandas as pd
    amazon_phones=pd.DataFrame({})
    amazon_phones['products']=products
    amazon_phones['prices']=prices
    amazon_phones['images']=images
    amazon_phones['avg_ratings']=avg_ratings
    return(amazon_phones)
```

```python
# Phones under 20000
df=amazon('https://www.amazon.in/Mobile-Phone-Under-20000-Rupees/s?k=Mobile+Phone+Under+20000+Rupees')
df
```

# 8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. 
You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description

```python
def weather(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    periods=[]
    period=soup.find_all('p',class_='period-name')
    for i in period:
        periods.append(i.text)
    
    short_descriptions=[]
    short_description=soup.find_all('p',class_='short-desc')
    for i in short_description:
        short_descriptions.append(i.text)
        
    temps=[]
    temp=soup.find_all('p',class_='temp')
    for i in temp:
        temps.append(i.text)
        
    descriptions=[]
    description=soup.find_all('div',class_='col-sm-10 forecast-text')
    for i in description:
        descriptions.append(i.text)
    descriptions=descriptions[:9]
    
    import pandas as pd
    weather=pd.DataFrame({})
    weather['periods']=periods
    weather['short_descriptions']=short_descriptions
    weather['temps']=temps
    weather['descriptions']=descriptions
    return(weather)
```

```python
# San Francisco Weather
df=weather('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.X80BdlUzbIU')
df
```

# 9. Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

```python
def monster(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    jobs=[]
    job=soup.find_all('h2',class_='title')
    for i in job:
        jobs.append(i.text.replace('\r\n',''))
    
    Days=[]
    day=soup.find_all('div',class_='meta flex-col')
    for i in day:
        Days.append(i.text.replace('\n','').replace('Applied\n\n\n\nSaved\n\n',''))
    
    Company=[]
    comp=soup.find_all('span',class_='name')
    # Even Indices are companies
    for i in range(0,len(comp),2):
        Company.append(comp[i].text)
        
    Location=[]
    # Odd Indices are companies
    for i in range(1,len(comp),2):
        Location.append(comp[i].text.replace('\r\n',''))
        
    import pandas as pd
    monster=pd.DataFrame({})
    monster['jobs']=jobs
    monster['Days']=Days
    monster['Company']=Company
    monster['Location']=Location
    return(monster)
```

```python
# Monster
df1=monster('https://www.monster.com/jobs/search/?q=Software-Developer&where=india')
df1
# Pages can be entered upto as many pages as required
```
